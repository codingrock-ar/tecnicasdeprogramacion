Estructuras condicionales

```python
# 1. Condicional if
valor1 = int(input("Ingrese un valor:"))

if valor1 > 0:
  print("El valor es positivo")
elif valor1 < 0:
  print("El valor es negativo")
else:
  print("El valor es cero")

# 2. Condicional if anidado
valor1 = int(input("Ingrese un valor:"))

if valor1 > 0:
  print("El valor es positivo")
  if valor1 % 2 == 0:
    print("El valor es par")
  else:
    print("El valor es impar")
elif valor1 < 0:
  print("El valor es negativo")
else:
  print("El valor es cero")

# 3. Condicional if con operador ternario
valor1 = int(input("Ingrese un valor:"))

resultado = "El valor es positivo" if valor1 > 0 else "El valor es negativo o cero"

print(resultado)
```

Uso de Switch

```python
# Python doesn't have a built-in switch statement, but you can use a dictionary to achieve the same functionality.

def switch_case(value):
  switcher = {
    1: "One",
    2: "Two",
    3: "Three",
    4: "Four",
    5: "Five",
    6: "Six",
    7: "Seven",
    8: "Eight",
    9: "Nine",
    10: "Ten"
  }
  return switcher.get(value, "Invalid value")

value = int(input("Enter a number between 1 and 10: "))
print(switch_case(value))
```

Estructuras de repeticiÃ³n

```python
# 1. While loop
valor1 = int(input("Ingrese un valor:"))

while valor1 > 0:
  print("El valor es positivo")
  valor1 -= 1

# 2. For loop
for i in range(1, 11):
  print(i)

# 3. Nested loops
for i in range(1, 11):
  for j in range(1, 11):
    print(i * j)

# 4. Break and continue statements
for i in range(1, 11):
  if i == 5:
    break
  print(i)

for i in range(1, 11):
  if i % 2 == 0:
    continue
  print(i)
```

```python
# 1. For loop
# Iterate over a list
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
  print(fruit)

# Iterate over a range of numbers
for i in range(1, 6):
  print(i)

# Iterate over a string
for letter in "banana":
  print(letter)

# 2. While loop
# Print numbers until the user enters a negative number
while True:
  num = int(input("Enter a number: "))
  if num < 0:
    break
  print(num)

# 3. Nested loops
# Print a multiplication table
for i in range(1, 11):
  for j in range(1, 11):
    print(i * j, end=" ")
  print()

# 4. Loop control statements
# Break statement
for i in range(1, 11):
  if i == 5:
    break
  print(i)

# Continue statement
for i in range(1, 11):
  if i == 5:
    continue
  print(i)

# Pass statement
for i in range(1, 11):
  if i == 5:
    pass  # Do nothing
  else:
    print(i)
```

```python
# 1. For loop

# Iterate over a list
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
  print(fruit)

# Iterate over a string
for letter in "banana":
  print(letter)

# Iterate over a range of numbers
for i in range(1, 6):
  print(i)

# 2. While loop

# Print numbers until the user enters a negative number
num = 0
while num >= 0:
  print(num)
  num = int(input("Enter a number: "))

# 3. Nested loops

# Print a multiplication table
for i in range(1, 11):
  for j in range(1, 11):
    print(i * j, end=" ")
  print()

# 4. Break and continue statements

# Break out of a loop
for i in range(1, 6):
  if i == 3:
    break
  print(i)

# Continue to the next iteration of a loop
for i in range(1, 6):
  if i == 3:
    continue
  print(i)

# 5. Pass statement

# Do nothing in a loop
for i in range(1, 6):
  pass
```

```python
# 1. While loop
# Initialize the variable
i = 1
n = int(input("Enter a positive integer: "))

# While loop to print numbers from 1 to n
while i <= n:
    print(i, end=" ")
    i += 1

# 2. For loop
# Iterate over a list
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
  print(fruit)

# Iterate over a range of numbers
for i in range(1, 6):
  print(i)

# Iterate over a range of numbers with a step
for i in range(1, 11, 2):
  print(i)

# 3. Nested loops
# Iterate over two lists
fruits = ["apple", "banana", "cherry"]
colors = ["red", "yellow", "green"]

for fruit in fruits:
  for color in colors:
    print(fruit, color)

# Iterate over a range of numbers with two nested loops
for i in range(1, 4):
  for j in range(1, 4):
    print(i, j)
```

Listas, arrays, tuplas, diccionarios, colas

```python
# prompt: show me how to use Listas, arrays, tuplas, diccionarios, colas, stacks, etc in python

# Listas
# Create a list
my_list = [1, 2, 3, 4, 5]

# Access elements of a list
print(my_list[0])  # Output: 1
print(my_list[-1])  # Output: 5

# Add elements to a list
my_list.append(6)
my_list.insert(1, 2.5)

# Remove elements from a list
my_list.remove(2.5)
my_list.pop()  # Remove the last element

# Iterate over a list
for element in my_list:
  print(element)

# Arrays
# Create an array
import numpy as np
my_array = np.array([1, 2, 3, 4, 5])

# Access elements of an array
print(my_array[0])  # Output: 1
print(my_array[-1])  # Output: 5

# Add elements to an array
my_array = np.append(my_array, 6)
my_array = np.insert(my_array, 1, 2.5)

# Remove elements from an array
my_array = np.delete(my_array, 1)
my_array = my_array[:-1]  # Remove the last element

# Iterate over an array
for element in my_array:
  print(element)

# Tuplas
# Create a tuple
my_tuple = (1, 2, 3, 4, 5)

# Access elements of a tuple
print(my_tuple[0])  # Output: 1
print(my_tuple[-1])  # Output: 5

# Add elements to a tuple
my_tuple = my_tuple + (6,)

# Remove elements from a tuple
# Tuples are immutable, so you cannot remove elements from them

# Iterate over a tuple
for element in my_tuple:
  print(element)

# Diccionarios
# Create a dictionary
my_dict = {"name": "John", "age": 30, "city": "New York"}

# Access elements of a dictionary
print(my_dict["name"])  # Output: John

# Add elements to a dictionary
my_dict["job"] = "Engineer"

# Remove elements from a dictionary
del my_dict["age"]

# Iterate over a dictionary
for key, value in my_dict.items():
  print(key, value)

# Colas
# Create a queue
from collections import deque
my_queue = deque([1, 2, 3, 4, 5])

# Add elements to a queue
my_queue.append(6)

# Remove elements from a queue
my_queue.popleft()  # Remove the first element

# Iterate over a queue
for element in my_queue:
  print(element)

# Stacks
# Create a stack
my_stack = []

# Add elements to a stack
my_stack.append(1)
my_stack.append(2)
my_stack.append(3)

# Remove elements from a stack
my_stack.pop()  # Remove the last element

# Iterate over a stack
for element in my_stack:
  print(element)
```
